##**Truth dataset**

##**Pre-processing**

```python
dp=pd.DataFrame(df_test.groupby('id')['cycle'].max()).reset_index()
dp.columns=['id','max']
dp
```

```python
df_truth['rtf']=df_truth['more']+dp['max']
df_truth
```

##**Calculate time to failure**

```python
df_truth.drop("more",axis="columns",inplace=True)
df_truth
```

```python
df_test=df_test.merge(df_truth,on=['id'],how='left')
df_test["ttf"]=df_test['rtf']-df_test['cycle']
df_test.drop("rtf",axis="columns",inplace=True)
df_test
```

```python
df_train['ttf']=df_train.groupby(['id'])['cycle'].transform(max)-df_train['cycle']
df_train
```

##**Applying threshold**

```python
period=30
df_train['label_bc']=df_train['ttf'].apply(lambda x: 1 if x <= period else 0)
df_test['label_bc']=df_test['ttf'].apply(lambda x: 1 if x <= period else 0)
df_train
```
